### Question 1


**What are the earliest, most recent, most common year of birth (only available for NYC and Chicago)?**

```r
# Function to find earliest birth year 
find.min.birthyear = function(table){
    result = min(table$Birth.Year, na.rm = TRUE)
    return(result)
}

# Function to find latest birth year
find.max.birthyear = function(table){
    result = max(table$Birth.Year, na.rm = TRUE)
    return(result)
}

# Function to find most common birth year
find.av.birthyear = function(table){
    result = mean(table$Birth.Year, na.rm = TRUE)
    return(result)
}
```

```r
# Displaying the earliest, most recent, and most common birth years from the New York table

ny.min.birthyear = find.min.birthyear(ny)
ny.max.birthyear = find.max.birthyear(ny)
ny.av.birthyear = find.av.birthyear(ny)
paste('New York earliest birth year, most recent, and most common:', round(ny.min.birthyear, digits=2),',', round(ny.max.birthyear, digits=2),',',round(ny.av.birthyear, digits=0))
```

```r
# Displaying the earliest, most recent, and most common birth years from the Chicago table 

chi.min.birthyear = find.min.birthyear(chi)
chi.max.birthyear = find.max.birthyear(chi)
chi.av.birthyear = find.av.birthyear(chi)
paste('Chicago earliest birth year, most recent, and most common:', round(chi.min.birthyear, digits=2),',', round(chi.max.birthyear, digits=2),',',round(chi.av.birthyear, digits=0))
```

```r
# Finding average age
ny.age = round(2017 - ny.av.birthyear, digits=0)
chi.age = round(2017 - chi.av.birthyear, digits=0)
paste('The average age for the bikers in the New York table is:', ny.age, ', while the average age for Chicago is:', chi.age)
```

**The earliest birth years were found through finding the minimum birth years in each of the cities. The most recent birth years were found by searching for the maximum birth years in each city, while the most common birth years were found through searching the mean birth year in each city. Based on the data found above and shown below, we can see that the earliest birth years for New York and Chicago as being the years 1885 and 1899. Since these people would be over a hundred years old, I am guessing that this is data that needs to be cleaned a little bit more. The most recent birth years in New York and Chicago are the years 2001 and 2002. While the most common birth years that use this biking system for travel in these two cities are 1978 and 1981. This means that the most common age for rental bikers in New York is 39, while the most common in Chicago is 36.**

```r
# Plotting the data
library(ggplot2)

# Creating the New York Birth Years Graph
ggplot(ny, aes(x = Birth.Year)) +
  geom_histogram(binwidth=2, color='blue', fill='light blue') +
  scale_x_continuous(limits = c(1885,2005)) +
  xlab('Birth Year') +
  ylab('Number of Users') +
  ggtitle('Bike User\'s Birth Years from New York')

# Creating the Chicago Birth Years Graph
ggplot(chi, aes(x = Birth.Year)) +
  geom_histogram(binwidth=2, color='black', fill='light green') +
  scale_x_continuous(limits = c(1885,2005)) +
  xlab('Birth Year') +
  ylab('Number of Users')+
  ggtitle('Bike User\'s Birth Years from Chicago')
```

### Question 2

**What is the average travel time for users in different cities?**

```r
# Function to display average travel time with different tables as input
find.mean = function(table){
    result = mean(table$Trip.Duration, na.rm = TRUE)
    return(result)
}

# Function to convert seconds to minutes
find.minutes = function(seconds){
    minutes = seconds / 60
    return(minutes)
}
```

```r
# Displaying average travel time for users in New York in minutes

ny.mean = find.mean(ny)
ny.minutes = find.minutes(ny.mean)
paste('New York average travel time:', round(ny.minutes, digits=2), 'minutes')
```

```r
# Displaying average travel time for users in Washington in minutes

wash.mean = find.mean(wash)
wash.minutes = find.minutes(wash.mean)
paste('Washington average travel time:', round(wash.minutes, digits=2), 'minutes')
```

```r
# Displaying average travel time for users in Chicago in minutes

chi.mean = find.mean(chi)
chi.minutes = find.minutes(chi.mean)
paste('Chicago average travel time:', round(chi.minutes, digits=2), 'minutes')
```

**The average travel times were found through searching the mean of each city's travel times. The average bike travel time for the city of New York is 15.06 minutes, Chicago's average is a little bit higher with an average of 15.62 minutes, and the city of Washington D.C. has the greatest biking average travel time with 20.57 minutes. This shows that in these three cities, the most common biking time is within 15-20 minutes. With these insights, it would be interesting to calculate the size of each city and whether that has an impact on the biking travel times.**

```r
# Plotting the data

# Creating the New York bike travel time graph
ggplot(ny, aes(x = Trip.Duration)) +
  geom_histogram(binwidth=1, color='blue') +
  scale_x_continuous(limits = c(0, 1000)) +
  xlab('Travel Time (seconds)') +
  ylab('Number of Users') +
  ggtitle('Bike Travel Times from New York')

# Creating the Washington D.C. bike travel time graph
ggplot(wash, aes(x = Trip.Duration)) +
  geom_histogram(binwidth=1, color='orange') +
  scale_x_continuous(limits = c(0, 1000)) +
  xlab('Travel Time (seconds)') +
  ylab('Number of Users') +
  ggtitle('Bike Travel Times from Washington D.C.')

# Creating the Chicago bike travel time graph
ggplot(chi, aes(x = Trip.Duration)) +
  geom_histogram(binwidth=1, color='green') +
  scale_x_continuous(limits = c(0, 1000)) +
  xlab('Travel Time (seconds)') +
  ylab('Number of Users') +
  ggtitle('Bike Travel Times from Chicago')
```

### Question 3

**What are the counts of each gender (only available for NYC and Chicago)?**

```r
# Function to return female count
female.count = function(table){
    count = nrow(subset(table, table$Gender == 'Female'))
    return(count)
}

# Function to return male count
male.count = function(table){
    count = nrow(subset(table, table$Gender == 'Male'))
    return(count)
}
```

```r
# Finding the gender counts of New York

ny.females = female.count(ny)
ny.males = male.count(ny)
paste('Number of females in New York table:', ny.females, ', number of males in New York table:', ny.males)
```

```r
# Finding the gender counts of Chicago

chi.females = female.count(chi)
chi.males = male.count(chi)
paste('Number of females in Chicago table:', chi.females, ', number of males in Chicago table:', chi.males)
```

**The numbers of each gender within each city were found through counting the number of rows where gender was equal to either male or female. We can see from the data displayed above that the number of females that rented bikes in New York was 12,159, while the number of males was 37,201. Therefore, we can see that males are more likely to rent a bicycle in New York than females. In Chicago, the number of females that rented a bike was 1,723, while the number of males was 5,159. Chicago also had a greater number of bike rentals from males, however, New York has a higher number of rentals in total.**

```r
# Plotting the data

# Creating the gender counts graph of New York
ggplot(data=subset(ny, !is.na(Gender)), aes(x = Gender, fill=Gender)) +
  geom_bar() +
  xlab('Gender') +
  ylab('Number of Users') +
  ggtitle('Number of Users by Gender in New York')

# Creating the gender counts graph of Chicago
ggplot(data=subset(chi, !is.na(Gender)), aes(x = Gender, fill=Gender)) +
  geom_bar() +
  xlab('Gender') +
  ylab('Number of Users') +
  ggtitle('Number of Users by Gender in Chicago')
```
